## Label Distribution

```python
def label_distribution(df, label_column='label'):
    
    label_counts = df[label_column].value_counts()
    label_percentages = (label_counts / label_counts.sum() * 100).round(2)

    label_distribution = pd.DataFrame({
        'Label': label_counts.index,
        'Count': label_counts.values,
        'Percentage': label_percentages.values
    })

    # Sort by label in ascending order
    label_distribution = label_distribution.sort_values(by='Label').reset_index(drop=True)

    return label_distribution
```

```python
# Parameters
target_total = 25000
num_classes = 15
current_size = len(pfc_df)
samples_needed = target_total - current_size
print(f"Need to sample {samples_needed} more rows from MFC")

# Current label distribution
pfc_counts = pfc_df['label'].value_counts().sort_index()

# Ideal class size
ideal_per_class = target_total // num_classes

# Track how many more per class we want (but don’t touch those already >= ideal)
samples_per_class_needed = {
    label: max(0, ideal_per_class - pfc_counts.get(label, 0))
    for label in range(num_classes)
}

# Collect needed samples from MFC
samples = []
used_indices = []

for label, n_needed in samples_per_class_needed.items():
    mfc_subset = mfc_df[mfc_df['label'] == label]
    n_available = len(mfc_subset)
    n_to_sample = min(n_needed, n_available)

    if n_to_sample > 0:
        sampled = mfc_subset.sample(n=n_to_sample, random_state=42)
        used_indices.extend(sampled.index)
        samples.append(sampled)

sampled_mfc_df = pd.concat(samples, ignore_index=True)

# Check how many more we need to hit 25k
total_so_far = len(pfc_df) + len(sampled_mfc_df)
filler_needed = target_total - total_so_far

if filler_needed > 0:
    unused_mfc = mfc_df.drop(index=used_indices, errors='ignore')
    filler = unused_mfc.sample(n=filler_needed, random_state=42, replace=True)
    sampled_mfc_df = pd.concat([sampled_mfc_df, filler], ignore_index=True)

# Final combined dataset
augmented_df = pd.concat([pfc_df, sampled_mfc_df], ignore_index=True)

# Report
# print(f"\nFinal dataset size: {len(augmented_df)}")
# print("Final label distribution (%):")
# print(augmented_df['label'].value_counts(normalize=True).sort_index().round(4) * 100)

```
