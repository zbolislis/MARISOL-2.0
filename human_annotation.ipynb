```python
import pandas as pd
import numpy as np
from statsmodels.stats import inter_rater as irr
```

```python
a_file = 'datasets/annotated/'
a1_file = f'{a_file}annotation_1.csv'
a2_file = f'{a_file}annotation_2.csv'
a3_file = f'{a_file}annotation_3.csv'
```

```python
# https://saturncloud.io/blog/how-to-remove-space-from-columns-in-pandas-a-data-scientists-guide/#:~:text=Using%20the%20str.strip(),column%20names%20or%20column%20values.
a1 = pd.read_csv(a1_file, usecols=['Text','Annotate'])
a1.rename(columns = {'Annotate':'Annotation-1'}, inplace = True)
a1['Text'] = a1['Text'].apply(lambda x: x.strip())
a1
```

```python
a2 = pd.read_csv(a2_file, usecols=['Text','Annotate'])
a2.rename(columns = {'Annotate':'Annotation-2'}, inplace = True)
a2['Text'] = a2['Text'].apply(lambda x: x.strip())
a2
```

```python
a3 = pd.read_csv(a3_file, usecols=['Text','Annotate'])
a3.rename(columns = {'Annotate':'Annotation-3'}, inplace = True)
a3['Text'] = a3['Text'].apply(lambda x: x.strip())
a3
```

```python
concated_a = pd.concat([a1, a2, a3], axis=1)
concated_a
```

```python
concatenated_a = concated_a.T.drop_duplicates().T
concatenated_a
```

```python
# https://stackoverflow.com/a/43990682
def getMajority(df, a, b, c):
    mostCommonVote=[]
    for row in df[[a, b, c]].values:
        votes, values = np.unique(row, return_counts=True)
        if np.all(values<=1):
                mostCommonVote.append(False)
        else:
            mostCommonVote.append(votes[np.argmax(values)])

    df['Majority-Frame'] = mostCommonVote
```

```python
getMajority(concatenated_a, 'Annotation-1', 'Annotation-2', 'Annotation-3')
concatenated_a
```

```python
concatenated_a['Majority-Frame'].value_counts()[False]
```

```python
output_path = f'{a_file}combined_HA_majority.csv'
concatenated_a.to_csv(f'{a_file}combined_HA_majority.csv', index=False)

print(f"Combined annotations saved to: {output_path}")
```

## Fleiss Kappa

```python
annotations = concatenated_a.copy()
annotations = annotations[['Annotation-1','Annotation-2','Annotation-3']]
annotations
```

```python
annotations_list = [annotations['Annotation-1'].tolist(), annotations['Annotation-2'].tolist(), annotations['Annotation-3'].tolist()]
```

```python
giro = np.array(annotations_list).transpose()
giro
```

```python
irr.fleiss_kappa(irr.aggregate_raters(giro)[0], method='fleiss')
```

## Encoding Code Frames

```python
ca_file = f'{a_file}combined_HA_complete.csv'
ca = pd.read_csv(ca_file)
ca
```

```python
ca_transformed = ca.copy()
```

```python
# Extract the numeric code from 'Majority-Frame'
ca_transformed['code_frames'] = ca_transformed['Majority-Frame'].str.extract(r'^(\d+)', expand=False).astype(int)

# Create the label column as code_frames - 1
ca_transformed['label'] = ca_transformed['code_frames'] - 1

# Keep only the desired columns
ca_transformed = ca_transformed[['Text', 'code_frames', 'label']].copy()

# Rename 'Text' to 'text'
ca_transformed.rename(columns={'Text': 'text'}, inplace=True)
```

```python
ca_transformed
```

```python
ca_output_path = f'{a_file}human_annotation_2000.csv'
ca_transformed.to_csv(ca_output_path, index=False)

print(f"Encoded annotations saved to: {ca_output_path}")
```
